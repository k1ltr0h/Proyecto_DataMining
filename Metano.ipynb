```python
import pandas as pd

ch4 = pd.read_csv("metano.csv", sep=',', index_col=0)
ch4
```

```python
cols = ch4.columns

new_ch4 = ch4.drop(["Year.1"], axis=1)

for x in range(len(cols)):
    new_ch4 = new_ch4.rename({cols[x]: "CH4 in "+str(cols[x])}, axis=1)
new_ch4
```

```python
old_data = pd.read_csv("p_p_g_g_a_e_f_s_a.csv", sep=',', index_col=0).drop(["index"],axis=1)
old_data
```

```python
years = new_ch4.index
final_ch4 = pd.concat([old_data.reset_index(), new_ch4.reset_index()], axis=1)
final_ch4 = final_ch4.set_index([years])
final_ch4 = final_ch4.drop(["Year"], axis=1)
display(final_ch4)
```

```python
final_ch4.to_csv('p_p_g_g_a_e_f_s_a_ch4.csv', index=True)
```
