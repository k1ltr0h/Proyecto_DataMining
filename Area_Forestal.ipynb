```python
import pandas as pd

forest = pd.read_csv('FAOSTAT_forest_land.csv', sep=',', index_col=0)
forest = forest.drop(["Domain", "Area Code", "Element Code", "Element", "Item Code", "Item", "Flag", "Flag Description"], axis=1)
forest = forest.set_index(["Year"])
display(forest)
```

```python
paises = forest["Area"].value_counts().index
pais_data = forest["Area"].value_counts().values

max_val = pais_data.max()

p70 = max_val*7/10

del_pais = []

for x in range(len(paises)):
    if(pais_data[x] < p70):
        del_pais.append(paises[x])
print(del_pais)

forest = forest.loc[~forest["Area"].isin(del_pais)]


display(forest)
#No hay nans
```

```python
forest_cols = forest["Area"].value_counts().index
years = forest["Year Code"].drop_duplicates().values

new_years = []

for i in years:
    if i >= 1998 and i <= 2014:
        new_years.append(i)
print(new_years)

data = pd.DataFrame()

prod = []

for year in new_years:
    #print(year)
    prod.append(forest.loc[forest["Year Code"] == year]["Value"].sum())
    #print(prod)
data = pd.concat([data, pd.DataFrame(prod, columns=["Forest_area[Kha]"])], axis=1).set_index([new_years])

display(data)
```

```python
old_data = pd.read_csv('p_p_g_a_e.csv', sep=',', index_col=0)
display(old_data)
```

```python
final_forest = pd.concat([old_data.reset_index(), data.reset_index()], axis=1)
final_forest = final_forest.set_index([new_years])
final_forest = final_forest.drop(["index"], axis=1)
display(final_forest)
```

```python
final_forest.to_csv('p_p_g_a_e_f.csv', index=True)
```
