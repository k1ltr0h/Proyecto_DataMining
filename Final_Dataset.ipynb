```python
import pandas as pd

old_data = pd.read_csv("p_p_g_g_a_e_f_s_a_ch4_no2.csv", sep=',', index_col=0)
old_data
```

```python
co2 = pd.read_csv("CO2_global.csv", sep=',', index_col=0)
co2
```

```python
import numpy as np

years = co2.index
cols = co2.columns
new_years = []
new_co2 = co2.reset_index()

for x in cols:
    if x == "World" or x == "Mundo":
        print("si existe")

for year in years:
    if year >= 1998 and year <= 2014:
        new_years.append(year)
print(new_years)
new_co2 = new_co2.loc[new_co2["Year"].isin(new_years)]
display(new_co2)
```

```python
max_val = len(new_years)
cols = new_co2.columns
pais_nans = new_co2.isna().sum()
print(pais_nans)

p30 = max_val*3/10

del_pais = []

for x in range(len(cols)):
    if(pais_nans[x] > p30):
        del_pais.append(cols[x])
        print(cols[x], x)
    else:
        new_co2 = new_co2.replace({cols[x] : {np.nan: new_co2[cols[x]].mean()}})
print(del_pais)
# Mt -> t*10^6
final_co2 = new_co2.drop(del_pais, axis=1).set_index(["Year"]).sum(axis=1).mul(1000000).rename("CO2 Total (tonnes)")
display(final_co2)
```

```python
years = final_co2.index
final_co2 = pd.concat([old_data.reset_index(), final_co2.reset_index()], axis=1)
final_co2 = final_co2.set_index([years])
final_co2 = final_co2.drop(["Year"], axis=1)
display(final_co2)
```

```python
final_co2.to_csv('Final_Dataset.csv', index=True)
```
