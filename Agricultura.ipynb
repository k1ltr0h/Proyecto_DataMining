```python
import pandas as pd

animales = pd.read_csv('animals.csv', sep=',', index_col=0)
display(animales)
```

```python
agricultura = pd.read_csv("FAOSTAT_agricultura.csv", sep=',')
display(agricultura)
```

```python
agricultura = agricultura.loc[agricultura["Año"] >= 1998]
display(agricultura)
```

```python
#print(agricultura["Unidad"].value_counts())
#print(agricultura["Ámbito"].value_counts())
#print(agricultura["Producto"].value_counts())

#display(agricultura.loc[agricultura["Producto"] == "Tierr. agrícolas bajo cubiertas protectoras"])
#print(agricultura["Área"].value_counts())

paises = agricultura["Área"].value_counts().index
pais_data = agricultura["Área"].value_counts().values

p50 = int(agricultura["Área"].value_counts().values.max())*50/100

#print(p50)
#print(paises, pais_data)

del_pais = []

for x in range(len(paises)):
    if(pais_data[x] < p50):
        del_pais.append(paises[x])
print(del_pais)

agricultura = agricultura.loc[~agricultura["Área"].isin(del_pais)]
```

```python
agricultura = agricultura.drop(["Código Ámbito", "Código área", "Ámbito", "Área", "Elemento", "Unidad", "Código año", "Código Producto", "Código Elemento", "Símbolo", "Descripción del Símbolo"], axis=1)

agricultura.info()

agricultura.rename(columns={'Valor':'Kha'}, inplace=True)


display(agricultura)

print(agricultura["Producto"].value_counts())
```

```python
agro_cols = agricultura["Producto"].value_counts().index
years = sorted(agricultura["Año"].value_counts().index)
print(agro_cols)
print(years)
data = pd.DataFrame()

for x in agro_cols:
    prod = []
    for year in years:
        prod.append(agricultura.loc[(agricultura["Producto"] == x) & (agricultura["Año"] == year)]["Kha"].sum())
    data = pd.concat([data, pd.DataFrame(prod, columns=[str(x) + "[Kha]"])], axis=1)
display(data)
```

```python
data = data.set_index([years])
gan_agri = pd.concat([data, animales], axis=1)
display(gan_agri)
```

```python
gan_agri.to_csv('ganaderia_agricultura.csv', index=True)
```
