```python
import warnings
warnings.filterwarnings('ignore') ## Desactivar Warnings
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv('Final_Dataset1_2.csv', sep=',', index_col=0)
display(data)
display(data.describe())
print(data.info())
```

```python
features = data.columns
tmp_data = data.T.reset_index()

tmp = pd.DataFrame()

for x in features:
    if x.split(" ")[0] == "CH4":
        tmp = pd.concat([tmp, data.loc[:,[x]]], axis=1)

data = pd.concat([data, tmp.sum(axis=1).to_frame("CH4 Total (tonnes CO2eq)")], axis=1)
display(data)
```

```python
cols = data.columns
std_data = StandardScaler().fit_transform(data)
std_data = pd.DataFrame(std_data, columns=cols)
display(std_data)
display(std_data.describe())
print(std_data.info())
```

```python
figure, heat = plt.subplots(figsize =(50,50))
data_corr = std_data.corr()#[["Codigo region", "Codigo comuna", "Poblacion", "Tasa", "Promedio"]]
labels = data_corr.index.values
heat = sns.heatmap(data_corr,
            xticklabels=labels, yticklabels=labels, annot=True)
```

```python
labels = ["N02 Total including land use change and forestry (tonnes CO₂e)", "CO2 Total (tonnes)", "CH4 Total (tonnes CO2eq)"]
sns.pairplot(std_data, y_vars=labels,x_vars=cols, hue="Population", height=4, palette=sns.cubehelix_palette(17))
```

```python
from sklearn.decomposition import PCA

y_labels = ["N02 Total including land use change and forestry (tonnes CO₂e)", "CO2 Total (tonnes)", "CH4 Total (tonnes CO2eq)"]
#isplay(std_data)
no2 = std_data.loc[:,[y_labels[0]]]
co2 = std_data.loc[:,[y_labels[1]]]
ch4 = std_data.loc[:,[y_labels[2]]]

new_std_data = std_data.drop(y_labels, axis=1)

pca = PCA(n_components=5)
principalComponents = pca.fit_transform(new_std_data)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc1', 'pc2', "pc3", "pc4", "pc5"])

print(principalDf.describe().loc[["std"]].values)
varianza = pow(principalDf.describe().loc[["std"]].values,2)
print(varianza[0])
#print(principalDf)
print(pca.explained_variance_ratio_)
fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.plot(pca.explained_variance_ratio_.cumsum())

langs = ["pc1","pc2","pc3","pc4","pc5"]
ax2.bar(langs, varianza[0])
plt.show()

pca_corr = principalDf.corr()
#print(pca_corr.max())
pca_corr = pca_corr
heat = sns.heatmap(pca_corr,
            xticklabels=pca_corr.columns.values, yticklabels=pca_corr.columns.values, annot=True)

new_principalDf = principalDf
new_principalDf[y_labels[0]] = no2
new_principalDf[y_labels[1]] = co2
new_principalDf[y_labels[2]] = ch4
#std_data
```

```python
#NO2
sns.pairplot(new_principalDf, y_vars=["pc1", "pc2", "pc3", "pc4", "pc5"],x_vars=["pc1", "pc2", "pc3", "pc4", "pc5"], hue=str(y_labels[0]), height=3, palette=sns.cubehelix_palette(17))
```

```python
#CO2
sns.pairplot(new_principalDf, y_vars=["pc1", "pc2", "pc3", "pc4", "pc5"],x_vars=["pc1", "pc2", "pc3", "pc4", "pc5"], hue=str(y_labels[1]), height=3, palette=sns.cubehelix_palette(17))
```

```python
#CH4
sns.pairplot(new_principalDf, y_vars=["pc1", "pc2", "pc3", "pc4", "pc5"],x_vars=["pc1", "pc2", "pc3", "pc4", "pc5"], hue=str(y_labels[2]), height=3, palette=sns.cubehelix_palette(17))
```

```python
from sklearn.cluster import KMeans

kmeans1 = KMeans(n_clusters=3, n_init=100, algorithm="elkan").fit(principalDf)
kmeans2 = KMeans(n_clusters=3, n_init=100, algorithm="full").fit(principalDf)
ctds1 = kmeans1.cluster_centers_
ctds2 = kmeans2.cluster_centers_
print(ctds1, "\n###########################\n", ctds2)

labels = kmeans.predict(principalDf)
colores=["r","c","y"]
pc = [[0,"pc1"],[1, "pc2"],[2, "pc3"],[3, "pc4"], [4, "pc5"]]
principalDf["K-Cluster"] = labels


def plot_kmeans(data, centroids):
    fig, ax = plt.subplots(nrows = len(pc), ncols = len(pc), figsize=(20,25))
    for i, name in pc:
        for u, u_name in pc:
            ax[i][u].set_title(name + " Vs " + u_name)
            ax[i][u].set_xlabel(name)
            ax[i][u].set_ylabel(u_name)
            ax[i][u].scatter(data[name], data[u_name], c=data["K-Cluster"], s=10)
            ax[i][u].scatter(centroids[:, i], centroids[:, u], marker='*', c=colores, s=500)
            
plot_kmeans(principalDf, ctds1)
plot_kmeans(principalDf, ctds2)
```

```python
k_1 = principalDf.loc[(principalDf["K-Cluster"] == 0)].loc[:,["pc1","pc2"]]
k_2 = principalDf.loc[principalDf["K-Cluster"] == 1].loc[:,["pc1","pc2"]]
k_3 = principalDf.loc[principalDf["K-Cluster"] == 2].loc[:,["pc1","pc2"]]
print(k_1,"\n Size: ",k_1.shape[0],"\n", k_2,"\n Size: ",k_2.shape[0],"\n", k_3,"\n Size: ",k_3.shape[0])
```
