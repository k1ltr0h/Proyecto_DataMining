```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('Final_Dataset.csv', sep=',', index_col=0)

display(data)
display(data.describe())
print(data.info())
```

```python
cols = data.columns
data = StandardScaler().fit_transform(data)
data = pd.DataFrame(data, columns=cols)
display(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

figure, heat = plt.subplots(figsize =(50,50))
data_corr = data.corr()#[["Codigo region", "Codigo comuna", "Poblacion", "Tasa", "Promedio"]]
labels = data_corr.index.values
heat = sns.heatmap(data_corr,
            xticklabels=labels, yticklabels=labels, annot=True)
```
