```python
import warnings
warnings.filterwarnings('ignore') ## Desactivar Warnings
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv('Final_Dataset1_3.csv', sep=',', index_col=0)
Years = list(data.index)
display(data)
display(data.describe())
print(data.info())
```

```python
features = data.columns
tmp_data = data.T.reset_index()

tmp = pd.DataFrame()

for x in features:
    if x.split(" ")[0] == "CH4":
        tmp = pd.concat([tmp, data.loc[:,[x]]], axis=1)

data = pd.concat([data, tmp.sum(axis=1).to_frame("CH4 Total (tonnes CO2eq)")], axis=1)
display(data)
```

```python
cols = data.columns
index = data.index
std_data = StandardScaler().fit_transform(data)
std_data = pd.DataFrame(std_data, columns=cols).set_index(index)
display(std_data)
display(std_data.describe())
print(std_data.info())
```

```python
from sklearn.decomposition import PCA

y_labels = ["N2O Total including land use change and forestry (tonnes CO₂e)", "CO2 Total (tonnes)", "CH4 Total (tonnes CO2eq)"]
#isplay(std_data)
n2o = std_data.loc[:,[y_labels[0]]]
co2 = std_data.loc[:,[y_labels[1]]]
ch4 = std_data.loc[:,[y_labels[2]]]
#display(n2o)

new_std_data = std_data.drop(y_labels, axis=1)
#display(n2o)
pca = PCA(n_components=5)
principalComponents = pca.fit_transform(new_std_data)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc1', 'pc2', "pc3", "pc4", "pc5"])

print(principalDf.describe().loc[["std"]].values)
varianza = pow(principalDf.describe().loc[["std"]].values,2)
print(varianza)
#print(principalDf)
print(pca.explained_variance_ratio_, pca.explained_variance_ratio_.sum())
fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.plot(pca.explained_variance_ratio_.cumsum())

langs = ["pc1","pc2","pc3","pc4","pc5"]
ax2.bar(langs, varianza[0])
plt.show()

pca_corr = principalDf.corr()
#print(pca_corr.max())
pca_corr = pca_corr
heat = sns.heatmap(pca_corr,
            xticklabels=pca_corr.columns.values, yticklabels=pca_corr.columns.values, annot=True)

new_principalDf = principalDf
new_principalDf
new_principalDf[y_labels[0]] = n2o.reset_index(drop=True)
new_principalDf[y_labels[1]] = co2.reset_index(drop=True)
new_principalDf[y_labels[2]] = ch4.reset_index(drop=True)
#std_data
display(new_principalDf)
```

```python
from sklearn.cluster import KMeans

kmeans1 = KMeans(n_clusters=3, n_init=100, algorithm="elkan").fit(principalDf)
kmeans2 = KMeans(n_clusters=3, n_init=100, algorithm="full").fit(principalDf)
ctds1 = kmeans1.cluster_centers_
ctds2 = kmeans2.cluster_centers_
print(ctds1, "\n###########################\n", ctds2)

labels = kmeans1.predict(principalDf)
colores=["r","c","y"]
pc = [[0,"pc1"],[1, "pc2"],[2, "pc3"],[3, "pc4"], [4, "pc5"]]
principalDf["K-Cluster"] = labels


def plot_kmeans(data, centroids):
    fig, ax = plt.subplots(nrows = len(pc), ncols = len(pc), figsize=(20,25))
    for i, name in pc:
        for u, u_name in pc:
            ax[i][u].set_title(name + " Vs " + u_name)
            ax[i][u].set_xlabel(name)
            ax[i][u].set_ylabel(u_name)
            ax[i][u].scatter(data[name], data[u_name], c=data["K-Cluster"], s=10)
            ax[i][u].scatter(centroids[:, i], centroids[:, u], marker='*', c=colores, s=500)
            
plot_kmeans(principalDf, ctds1)
plot_kmeans(principalDf, ctds2)
```

# Ridge y Lasso

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import mean_squared_error as mse
```

```python
ridge_rg = Ridge(alpha=1.0)
ridge_rg.fit(x_train_ch4, y_train_ch4)
yrid_predict_ch4_train = ridge_rg.predict(x_train_ch4)
yrid_predict_ch4_test = ridge_rg.predict(x_test_ch4)
```

```python
#Error train
errorRid_train = mse(y_train_ch4, yrid_predict_ch4_train)

#Error validación
errorRid_valid = mse(y_test_ch4, yrid_predict_ch4_test)

print('error MSE Ridge Train :', errorRid_train)
print('error MSE Ridge Train :', errorRid_valid)
```

```python

Y_train = []
Y_test = []
def model_plot_score(alfa, x_train, y_train, x_test, y_test, model):
    errorRid_train = []
    errorRid_valid = []
    train_score = []
    test_score = []
    diff = []
    for a in alfa:
        if model == 'Ridge':
            clf = Ridge(alpha = a)
        elif model == 'Lasso':
            clf = Lasso(alpha= a)
        else:
            clf = Ridge(alpha = a)
            
        clf.fit(x_train, y_train)
        yrid_predict_train = clf.predict(x_train)
        yrid_predict_test = clf.predict(x_test)
        
        train_score.append(clf.score(x_train, y_train))
        test_score.append(clf.score(x_test, y_test))
        et =float( mse(y_train, yrid_predict_train) )
        ev =float( mse(y_test, yrid_predict_test) )
        errorRid_train.append(et) 
        errorRid_valid.append(ev)
        diff.append(abs(et - ev))
    plt.figure(figsize=(16,12))
    plt.subplot(2,1,1)
    #plt.plot(alfa, train_score, color='red')
    #plt.plot(alfa, test_score, color='blue')
    #plt.scatter(alfa, train_score, color='red', label = 'Train Score')
    #plt.scatter(alfa, test_score, color='blue', label = 'Test Score')
    plt.scatter(alfa, errorRid_train, color='red', label = 'Train mse')
    plt.scatter(alfa, errorRid_valid, color='blue', label = 'Test mse')
    plt.xlabel('Alpha')
    plt.ylabel('MSE')
    plt.grid()
    plt.legend(fontsize = 20)
    plt.subplot(2,1,2)
    
    plt.bar(list(x_train.columns), clf.coef_)
    plt.ylabel('Coeficiente')
    plt.xticks(rotation='vertical')
    plt.title('Coeficientes modelo '+model)
    print("min mse :", min(diff))
    print('alpha: ', alfa[diff.index(min(diff))])

alfa = np.linspace(0,10, 20)
model_plot_score(alfa, x_train_ch4, y_train_ch4, x_test_ch4, y_test_ch4, 'Ridge')
```
