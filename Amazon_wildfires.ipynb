```python
import pandas as pd

amazon = pd.read_csv('amazon_wildfires.csv', sep=',', index_col=0, engine='python')

display(amazon)
```

```python
new_amazon = amazon.reset_index()
display(new_amazon)
print(sorted(new_amazon["year"].value_counts()))
```

```python
#amazon["date"].str.split("-")
amazon_cols = new_amazon["state"].value_counts().index
years = new_amazon["year"].drop_duplicates().values

new_years = []

for i in years:
    if i >= 1998 and i <= 2014:
        new_years.append(i)
print(new_years)

data = pd.DataFrame()

prod = []

for year in new_years:
    #print(year)
    prod.append(new_amazon.loc[new_amazon["year"] == year]["number"].sum())
    #print(prod)
data = pd.concat([data, pd.DataFrame(prod, columns=["N_incendios(Amazonía BR)"])], axis=1).set_index([new_years])

display(data)
```

```python
old_data = pd.read_csv('p_p_g_a_e_f_s.csv', sep=',', index_col=0)

final_amazonia = pd.concat([old_data.reset_index(), data.reset_index()], axis=1)
final_amazonia = final_amazonia.set_index([new_years])
final_amazonia = final_amazonia.drop(["index"], axis=1)
display(final_amazonia)
```

```python
final_amazonia.to_csv('p_p_g_a_e_f_s_a.csv', index=True)
```
