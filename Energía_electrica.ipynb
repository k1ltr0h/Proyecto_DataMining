```python
import pandas as pd
energia = pd.read_csv('Energía_electrica.csv', sep=',', index_col=0, skiprows=4).drop(["Indicator Name", "Indicator Code", "Country Code", "Unnamed: 64"], axis=1)

display(energia)
```

```python
new_energia = energia.T

print(new_energia.index)

del_year = []
for x in new_energia.index:
    if(int(x) < 1998 or int(x) > 2014):
        del_year.append(x)
new_energia = new_energia.drop(del_year)

#petro = petro.loc[petro["Country Name"] >= 1990]

display(new_energia)
```

```python
import numpy as np

nan_cols = new_energia.isna().sum()

n_years = len(new_energia.index)
cols = new_energia.columns
print(nan_cols)

del_cols = []

p30 = nan_cols.max()*3/10

for x in range(len(cols)):
    if(nan_cols[x] > p30):
        #rint(nan_cols[x])
        del_cols.append(cols[x])
    new_energia = new_energia.replace({cols[x] : {np.nan: new_energia[cols[x]].mean()}})
display(del_cols)
new_energia = new_energia.drop(del_cols, axis=1)
display(new_energia)    
```

```python
new_energia = new_energia.sum(axis=1)
new_energia = new_energia.to_frame("kWh per cápita")
display(new_energia)
```

```python
data = pd.read_csv('p_g_a.csv', sep=',', index_col=0)

years = new_energia.index

new_data = pd.concat([data.reset_index(), new_energia.reset_index()], axis=1)
new_data = new_data.set_index(years).drop(["index"], axis=1)
display(new_data)
```

```python
new_data.to_csv('p_g_a_e.csv', index=True)
```
