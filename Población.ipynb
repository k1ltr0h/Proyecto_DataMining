```python
import pandas as pd
pob = pd.read_csv('world-population.csv', sep=',', index_col=0).drop(["Code"], axis=1)

display(pob)
```

```python
import numpy as np

years = pob["Year"].drop_duplicates().values
n_years = len(years)
cols = pob.index.to_series().drop_duplicates().values

new_years = []

for x in years:
    try:
        if int(x) <= 2014 and int(x) >= 1998:
            new_years.append(x)
    except:
        print("Skip") 
print(cols)
print(new_years)


```

```python
year_col = "Year"

new_pob = pob.loc[pob[year_col].isin(new_years)].loc["World",:].set_index([new_years]).drop([year_col], axis=1)

display(new_pob)

```

```python
data = pd.read_csv('p_g_a_e.csv', sep=',', index_col=0)

years = new_pob.index

new_data = pd.concat([data.reset_index(), new_pob.reset_index()], axis=1)
new_data = new_data.set_index(years).drop(["index"], axis=1)

display(new_data)
```

```python
new_data.to_csv('p_p_g_a_e.csv', index=True)
```
