```python
import warnings
warnings.filterwarnings('ignore') ## Desactivar Warnings
import pandas as pd

co2 = pd.read_csv('co-emissions-per-capita.csv', sep=',', index_col=0)
old_data = pd.read_csv('Final_Dataset1_2.csv', sep=',', index_col=0).rename({"N02 Total including land use change and forestry (tonnes CO₂e)": "N2O Total including land use change and forestry (tonnes CO₂e)"}, axis=1)
```

```python
co2 = co2.drop(["Code"], axis=1)
display(co2)
```

```python
import numpy as np

years = co2["Year"].drop_duplicates().values
n_years = len(years)
cols = co2.index.to_series().drop_duplicates().values

new_years = []

for x in years:
    try:
        if int(x) <= 2014 and int(x) >= 1998:
            new_years.append(x)
    except:
        print("Skip") 
print(cols)
print(new_years)
```

```python
new_co2 = co2.loc[co2["Year"].isin(new_years)].T["World"].T.reset_index().drop(["Entity", "Year"], axis=1)
display(new_co2)
```

```python
#pd.concat([old_data["Population"].reset_index(), new_co2], axis=1)
#display(old_data["Population"])
new_co2 = (old_data["Population"].reset_index().drop(["Year"],axis=1).rename({"Population":"CO2 Total (tonnes)"}, axis=1)).mul(new_co2.rename({"Per capita CO₂ emissions (tonnes)":"CO2 Total (tonnes)"},axis=1))
display(new_co2)
```

```python
index = old_data.index
final_data = pd.concat([old_data.reset_index().drop(["Year", "CO2 Total (tonnes)"],axis=1), new_co2],axis=1).set_index(index)
display(final_data)
```

```python
final_data.to_csv('Final_Dataset1_3.csv', index=True)
```
