```python
import pandas as pd
petro = pd.read_csv('Petróleo.csv', sep=',', index_col=0)

petro = petro.drop(["Indicator Code", "Country Code", "Indicator Name"], axis=1)
# Uso de energía (kg de equivalente de petróleo per cápita)

petro = petro.T

print(petro.index)

del_year = []
for x in petro.index:
    if(int(x) < 1998 or int(x) > 2014):
        del_year.append(x)
petro = petro.drop(del_year)

#petro = petro.loc[petro["Country Name"] >= 1990]

display(petro)
```

```python
import numpy as np

nan_cols = petro.isna().sum()

n_years = len(petro.index)
cols = petro.columns
print(nan_cols)

del_cols = []

p30 = nan_cols.max()*3/10

for x in range(len(cols)):
    if(nan_cols[x] > p30):
        #rint(nan_cols[x])
        del_cols.append(cols[x])
    petro = petro.replace({cols[x] : {np.nan: petro[cols[x]].mean()}})
display(del_cols)
petro = petro.drop(del_cols, axis=1)
display(petro)    
```

```python
petro = petro.sum(axis=1)
```

```python
petro = petro.to_frame("kg eq de petróleo per cápita")
```

```python
gan_agri = pd.read_csv('ganaderia_agricultura.csv', sep=',', index_col=0)
display(gan_agri)
```

```python
years = gan_agri.index

final_petro = pd.concat([gan_agri.reset_index(), petro.reset_index()], axis=1)
final_petro = final_petro.set_index([years])
final_petro = final_petro.drop(["index"], axis=1)
display(final_petro)
```

```python
print(final_petro.index)

del_year = []
for x in final_petro.index:
    if(int(x) > 2014):
        del_year.append(x)
final_petro = final_petro.drop(del_year)

#petro = petro.loc[petro["Country Name"] >= 1990]

display(final_petro)
```
